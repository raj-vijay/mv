**Superpixels and Segmentation**

A single pixel in an image is not in anysense a real representation of the image. In order to derive logical meanings from the image, superpixels, which cover multiple pixels of the image, are used.

A superpixel is a group of connect pixels with a similar grade or colour. A superpixel is an image patch which is better aligned with intensity edges than a rectangular patch. Superpixels can be extracted with any segmentation algorithm, however, most of them produce highly irregular superpixels, with widely varying sizes and shapes. A more regular space tessellation may be desired.

**Benefits of Superpixels**

* More meaningful regions
* Computational efficiency

**Two types of Segmentation**

* Supervised - Prior knowledge is required, such as thresholding.
* Unsupervised

```
import matplotlib.pyplot as plt
import numpy as np
def show_image(image, title='Image', cmap_type='gray'):
  plt.imshow(image, cmap=cmap_type)
  plt.title(title)
  plt.axis('on')
  plt.show()
```

```
def plot_comparison(original, filtered, title_filtered):
  fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 6), sharex=True,
  sharey=True)
  ax1.imshow(original, cmap=plt.cm.gray)
  ax1.set_title('Original')
  ax1.axis('off')
  ax2.imshow(filtered, cmap=plt.cm.gray)
  ax2.set_title(title_filtered)
  ax2.axis('off')
```

**Unsupervised segmentation**

Unsupervised segmentation carried out using the **Simple Linear Iterative Clustering (SLIC)** algorithm.

```
from skimage import data
image = data.coffee()
```

```
# Import the modules
from skimage.segmentation import slic
from skimage.color import label2rgb
# Obtain the segments
segments = slic(image)
# Put segments on top of original image to compare
segmented_image = label2rgb(segments, image, kind='avg')
show_image(image)
show_image(segmented_image, "Segmented image")
```

**More Segments**

```
# Import the modules
from skimage.segmentation import slic
from skimage.color import label2rgb
# Obtain the segmentation with 300 regions
segments = slic(image, n_segments= 300)
# Put segments on top of original image to compare
segmented_image = label2rgb(segments, image, kind='avg')
show_image(segmented_image)
```

**Superpixel segmentation example**

```
face_image = plt.imread('person.jpg')
```

```
# Import the slic function from segmentation module
from skimage.segmentation import slic

# Import the label2rgb function from color module
from skimage.color import label2rgb

# Obtain the segmentation with 400 regions
segments = slic(face_image, n_segments = 400)

# Put segments on top of original image to compare
segmented_image = label2rgb(segments, face_image, kind='avg')

# Show the segmented image
show_image(face_image, 'Original')
show_image(segmented_image, "Segmented image, 400 superpixels")
```
