**Define function to show image**

```
import matplotlib.pyplot as plt
def show_image(image, title='Image', cmap_type='gray'):
  plt.imshow(image, cmap=cmap_type)
  plt.title(title)
  plt.axis('on')
  plt.show()
```

```
# Loading the image using Matplotlib
image = plt.imread('MV_01_image_1.jpg')
type(image)
```

**Colours with NumPy**

A colour image is an NumPy array with the third dimension as the colour channel, for RGB.

```
# Obtaining the red values of the image
red = image[:, :, 0]
# Obtaining the green values of the image
green = image[:, :, 1]
# Obtaining the blue values of the image
blue = image[:, :, 2]
```

```
plt.imshow(red, cmap="gray")
plt.title('Red')
plt.axis('off')
plt.show()
```

```
plt.imshow(green, cmap="gray")
plt.title('Green')
plt.axis('off')
plt.show()
```

```
plt.imshow(blue, cmap="gray")
plt.title('Blue')
plt.axis('off')
plt.show()
```

**Flipping images: vertically**

```
# Flip the image in up direction
vertically_flipped = np.flipud(image)
show_image(vertically_flipped, 'Vertically flipped image')
```

**Flipping images: horizontally**

```
# Flip the image in left direction
horizontally_flipped = np.fliplr(image)
show_image(horizontally_flipped, 'Horizontally flipped image')
```

**Image Histogram**

Image histogram is a graphical representation of intensity values of each pixel values. 

Applications of histograms
1. Analysis
2. Thresholding
3. Brightness and contrast
4. Equalize an image

```
# Red color of the image
red = image[:, :, 0]
# Obtain the red histogram
plt.hist(red.ravel(), bins=256)
plt.title('Red Histogram')
plt.show()
```

```
green = image[:, :, 1]
plt.hist(green.ravel(), bins=256)
plt.title('Green Histogram')
plt.show()
```

```
blue = image[:, :, 2]
plt.hist(blue.ravel(), bins=256)
plt.title('Blue Histogram')
plt.show()
```
